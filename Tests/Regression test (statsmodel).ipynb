```python
%matplotlib inline
import pandas as pd
import numpy as np
import statsmodels.api as sm
from sklearn import datasets
```

```python
data = datasets.load_boston()
print(data.DESCR)
print(data.feature_names)
print(data.target)
```

```python
# Independent variables
df = pd.DataFrame(data.data, columns = data.feature_names)
print(df)
```

```python
# Dependent variables that the regression model will predict
target = pd.DataFrame(data.target, columns = ["MEDV"])
print(target)
```

# Single variable regression

```python
# Correlation without a constant
X = df["RM"]
y = target["MEDV"]

# Generate a model using the ordinary least squares method
model = sm.OLS(y, X).fit()
predictions = model.predict(X) # Make the predictions by the model

# Print a summary of the model
model.summary()
```

```python
# Generate a model with correlation constants

X = df["RM"]
y = target["MEDV"]
X = sm.add_constant(X) # Add a y-intercept to the model

model = sm.OLS(y, X).fit()
predictions = model.predict(X)

model.summary()
```

# Multivariable regression

```python
X = df[["RM", "LSTAT"]]
y = target["MEDV"]

model = sm.OLS(y, X).fit()
predictions = model.predict(X)

model.summary()
```
